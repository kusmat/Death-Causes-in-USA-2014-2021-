# Analysis of Death causes and their count

### Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Getting data into csv files

```python
#Files are downloaded as csv files from publicly available sources: search on Data.gov.
DeathsAfter2020 = pd.read_csv('Monthly_Provisional_Counts_of_Deaths_by_Select_Causes__2020-2021.csv')
DeathsAfter2014 = pd.read_csv('Monthly_Counts_of_Deaths_by_Select_Causes__2014-2019.csv')
```

```python
DeathsAfter2014.head()
```

```python
DeathsAfter2020.head()
```

### Getting column names from both tables for comparison and then comparing both tables before merging

```python
DeathsAfter2020columns = pd.Series(DeathsAfter2020.columns, name = 'DeathsAfter2020columns')
DeathsAfter2014columns = pd.Series(DeathsAfter2014.columns, name = 'DeathsAfter2014columns')
```

```python
#below commented out for clarity, but used info() and print() to inspect data first
# DeathsAfter2014columns
# DeathsAfter2020.info()
# # DeathsAfter2020columns
# DeathsAfter2020.info()
```

```python
ColumnsCompareDf = pd.merge(DeathsAfter2014columns, DeathsAfter2020columns, right_index=True, left_index=True)
```

```python
#comparing both columns sets to understand the differences:
ColumnsCompareDf
```

### Merging both data sets as most of the needed columns are common to both.

```python
DeathsCausesAll = pd.concat([DeathsAfter2014, DeathsAfter2020], axis = 0, join = 'outer') #merging on columns
```

```python
DeathsCausesAll.info()
```

### Removing columns that are not needed and only pollute data for analysis

```python
DeathsCausesAllCleaned = pd.concat([DeathsCausesAll.iloc[:, 0:21], DeathsCausesAll.iloc[:, 24:28]], axis=1)
```

```python
DeathsCausesAllCleaned.info()
```

### Adding Date column, consisting of Year and Month columns and sorting data by this column

```python
# Our goal now is to have a date column and sort data by date:

    # With current data, we only have separate Year and Month columns, so we cannot sort by them.
    # We also have duplicate indexes so we cannot use them as is.
    # We also notice that data is not sorted chronologically but by running DeathsAfter2014.head().
    # We see that we have all data from January for each year, so this needs to be corrected.
```

```python
#checking data we need, we notice that because we combined both datasets, we have doubled indexes:
DeathsCausesAllCleaned['Year'][0]
```

```python
# resetting indexes
DeathsCausesAllCleaned.reset_index(inplace=True)
```

```python
#Trial of the formula, to see if conversion of the Year/Month to date will work.
pd.to_datetime(str(DeathsCausesAllCleaned['Year'][1])+"/"+str(DeathsCausesAllCleaned['Month'][1])+"/"+"1")
#voila!
```

```python
#Creating a series to be used as the new column in the dataframe as date.
# This is based on previous formula, but will do it for all rows.
YearMonth = []
for element in range(0, len(DeathsCausesAllCleaned)):
    YearMonth.append(pd.to_datetime(str(DeathsCausesAllCleaned['Year'][element])+"/"+str(DeathsCausesAllCleaned['Month'][element])+"/"+"1"))
```

```python
#Adding new column of data based on series just created:
DeathsCausesAllCleaned['YearMonth'] = YearMonth
```

```python
DeathsCausesAllCleaned.sort_values(by = 'YearMonth', axis = 0, ascending = True, inplace=True)
```

```python
#after sorting we have wrong indexes, so need to reset them:
#DeathsCausesAllCleaned.head()
```

```python
DeathsCausesAllCleaned.reset_index(inplace=True)
```

```python
DeathsCausesAllCleaned.head()
```

```python
#removing extra indexes:
DeathsCausesAllCleaned = DeathsCausesAllCleaned.drop(['level_0', 'index'], axis = 1)
```

```python
#DeathsCausesAllCleaned.tail()
```

```python
# After funning this plot command for general data inspection we notice that December 2021 data is not complete.
# Data from december weill need to be removed.
plt.figure(figsize=(20,10))
DeathsCausesAllCleaned.plot()
```

```python
DeathsCausesAllCleaned = DeathsCausesAllCleaned.iloc[0:-1, :]
```

```python
DeathsCausesAllCleaned.tail()
```

#### We also noticed that 2 columns are duplicated due to different naming in 2020/2021:

* Nephritis, Nephrotic Syndrome, and Nephrosis	
* Nephritis, Nephrotic Syndrome and Nephrosis

* Symptoms, Signs, and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified
* Symptoms, Signs and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified

#### We will need to merge them into one.



```python
NephritisData = DeathsCausesAllCleaned['Nephritis, Nephrotic Syndrome, and Nephrosis'].isnull().sum()
```

```python
NephritisData1 = DeathsCausesAllCleaned['Nephritis, Nephrotic Syndrome, and Nephrosis'].fillna(value = "0").astype(int)
NephritisData2 = DeathsCausesAllCleaned['Nephritis, Nephrotic Syndrome and Nephrosis'].fillna(value = "0").astype(int)
SymptomsData1 = DeathsCausesAllCleaned['Symptoms, Signs and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified'].fillna(value = "0").astype(int)
SymptomsData2 = DeathsCausesAllCleaned['Symptoms, Signs, and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified'].fillna(value = "0").astype(int)

def strToInt (value):    
    return int(value)

# DeathsCausesAllCleaned['Nephritis, Nephrotic Syndrome, and Nephrosis'] = strToInt(DeathsCausesAllCleaned['Nephritis, Nephrotic Syndrome, and Nephrosis'])
# DeathsCausesAllCleaned['Nephritis, Nephrotic Syndrome and Nephrosis'] = strToInt(DeathsCausesAllCleaned['Nephritis, Nephrotic Syndrome and Nephrosis'])
# DeathsCausesAllCleaned['Symptoms, Signs and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified'] = strToInt(DeathsCausesAllCleaned['Symptoms, Signs and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified'])
# DeathsCausesAllCleaned['Symptoms, Signs, and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified'] = strToInt(DeathsCausesAllCleaned['Symptoms, Signs, and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified'])
```

```python
NephritisData1
```

```python
NephritisData = NephritisData1 + NephritisData2
SymptomsData = SymptomsData1 + SymptomsData2
```

```python
DeathsCausesAllCleaned['Nephritis, Nephrotic Syndrome and Nephrosis'] = NephritisData
DeathsCausesAllCleaned.drop('Nephritis, Nephrotic Syndrome, and Nephrosis', axis = 1, inplace=True)
DeathsCausesAllCleaned['Symptoms, Signs and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified'] = SymptomsData
DeathsCausesAllCleaned.drop('Symptoms, Signs, and Abnormal Clinical and Laboratory Findings, Not Elsewhere Classified', axis = 1, inplace=True)
```

### Now the data cleaning is complete and we can create trends to understand the data

* First trend is the comparison on All causes and Natual causes together with covid

```python
plt.figure(figsize=(10,5))
plt.plot(DeathsCausesAllCleaned['YearMonth'], DeathsCausesAllCleaned['All Cause'])
plt.plot(DeathsCausesAllCleaned['YearMonth'], DeathsCausesAllCleaned['Natural Cause'])
plt.plot(DeathsCausesAllCleaned['YearMonth'], DeathsCausesAllCleaned['COVID-19 (Multiple Cause of Death)'])
plt.plot(DeathsCausesAllCleaned['YearMonth'], DeathsCausesAllCleaned['COVID-19 (Underlying Cause of Death)'])
#sns.lineplot(x=DeathsCausesAllCleaned['YearMonth'], y = DeathsCausesAllCleaned['Motor Vehicle Accidents'], hue=DeathsCausesAllCleaned['Jurisdiction of Occurrence'])
plt.axis()
```

* Second trend the comparison on Deaths with covid and without covid and check % covid in comparison to all data

```python
PercentCovidDeaths = (DeathsCausesAllCleaned['COVID-19 (Underlying Cause of Death)']/DeathsCausesAllCleaned['All Cause'])*100
```

```python
DeathsCausesAllCleaned['% Covid Deaths'] = PercentCovidDeaths
```

```python
DeathsWithoutCovid = DeathsCausesAllCleaned['All Cause'] - DeathsCausesAllCleaned['COVID-19 (Underlying Cause of Death)']
```

```python
DeathsCausesAllCleaned['All Cause less Covid'] = DeathsWithoutCovid
```
